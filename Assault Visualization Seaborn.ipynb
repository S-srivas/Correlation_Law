```python
# Today I'm plotting the actual crime data in which I got the Aggravated Assualts data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
from pylab import bar, show, plot
import seaborn as sns
import timeit

# This is what the file is called
df=pd.read_csv('Crime Data.csv')
```

```python
# This tells me the rows, columns
df.shape
```

```python
# By this I can see the different data types in the file
df.info()
```

```python
# This basically shows the Violent crime rate (VCR) plotted with the x axis being data points
# Type: Float64
sns.distplot(df.VCR.dropna())
plt.show()
```

```python
# This basically creates a heatmap representing all the data in the file
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# This was a heatmap for specific data points
df1 = df.pivot_table(index='VC', columns='RapeRate', values='RobberyRate', aggfunc=np.median)
sns.heatmap(df1, annot=True, fmt=".1f")
plt.show()
```

```python
# # This basically shows the Burglary rate plotted with the x axis being data points. Also works on frequency
# Type: Float64
sns.distplot(df.BurglaryRate.dropna())
plt.show()
```
